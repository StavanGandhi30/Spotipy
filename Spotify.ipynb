# Importing Spotify and Other Libraries

```python
import spotipy
from pprint import pprint
from spotipy.oauth2 import SpotifyOAuth, SpotifyClientCredentials
```

### Passing API

```python
client_id = "<Your Client Id>"
client_secret = "<Your Client Secret>"
redirect_uri = "<Your Redirect URI>"
scope = "user-library-read"
```

### Initializing Spotify

```python
sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=client_id,
                                               client_secret=client_secret,
                                               redirect_uri=redirect_uri,
                                               scope=scope))
```

### Creating Spotify Search

```python
spotifySearch = sp.search(q='hindi music', limit=20, type='track', market="IN")
for idx, track in enumerate(spotifySearch['tracks']['items']):
    print(idx+1, track['name'])
```

### Getting Artist's Top Tracks

```python
ArijitSingh_URI = 'spotify:artist:4YRxDV8wJFPHPTeXepOstw'
artist_top_tracks = sp.artist_top_tracks(ArijitSingh_URI)

for track in artist_top_tracks['tracks']:
    print('track    : ' + track['name'])
    print('audio    : ' + track['preview_url'], end="\n\n")
```

### Getting Newly Released Albums and Tracks 

```python
sp.new_releases(country="IN")['albums']['items'][0]['artists'][0]['name']

for album in sp.new_releases(country="IN", offset=0, limit=50)['albums']['items']:
    print("Artists:", album['artists'][0]['name'])
    print("Album:", album['name'], end="\n\n")
```

### Getting User Saved Tracks

```python
user_saved_tracks = sp.current_user_saved_tracks()
for idx, item in enumerate(user_saved_tracks['items']):
    track = item['track']
    print(idx+1, track['artists'][0]['name'], " – ", track['name'], end="\n\n")
```

### Getting User Playlists

```python
user_playlists = sp.user_playlists('31p7lpuwznxoewkiuernmxfcxb24')
playlists = ""
while playlists:
    for i, playlist in enumerate(playlists['items']):
        print("%4d %s %s" % (i + 1 + playlists['offset'], playlist['uri'],  playlist['name']))
    if playlists['next']:
        playlists = sp.next(playlists)
    else:
        playlists = None
```

### Getting Artist related to another Artist

```python
ArijitSingh_URI = 'spotify:artist:4YRxDV8wJFPHPTeXepOstw'
artist_related_artists = sp.artist_related_artists(ArijitSingh_URI)["artists"]

for artist in artist_related_artists:
    print(artist["name"], "-" , artist["uri"], end="\n\n")
```

### Getting Top Tracks of Artist related to another Artist

```python
ArijitSingh_URI = 'spotify:artist:4YRxDV8wJFPHPTeXepOstw'

artist_related_artists = sp.artist_related_artists(ArijitSingh_URI)["artists"]

for artist in artist_related_artists:
    print(artist["name"], "-" ,artist["uri"])
    for track in sp.artist_top_tracks(artist["uri"])['tracks']:
        print('track    : ' + track['name'])
        print('audio    : ' + track['preview_url'], end="\n\n")
```

### Getting Featured Playlists by Country and Locality

```python
sp.featured_playlists(locale="hi", country="IN", limit=20, offset=0)['playlists']['items']

for album in sp.featured_playlists(locale="hi", country="IN", limit=20, offset=0)['playlists']['items']:
    print("Display Name -", album['owner']['display_name'])
    print("Album Name -", album['name'])
    print("URI -", album['uri'], end="\n\n")
```

### Getting Song Recommendation using Artists

```python
recommendedArtist = [
    "spotify:artist:4YRxDV8wJFPHPTeXepOstw",
    "spotify:artist:4fEkbug6kZzzJ8eYX6Kbbp",
    "spotify:artist:7HCqGPJcQTyGJ2yqntbuyr",
    "spotify:artist:4oVMLzAqW6qhRpZWt8fNw4",
    "spotify:artist:2oSONSC9zQ4UonDKnLqksx"
]

musicRecommendations = sp.recommendations(seed_artists=recommendedArtist, limit=100, country="IN")

for tracks in musicRecommendations['tracks']:
    if tracks["popularity"] >= 70: 
        print("Album -", tracks["album"]["name"])
        print("Track -",tracks["name"])
        print("Popularity -",tracks["popularity"], end="\n\n")
```
